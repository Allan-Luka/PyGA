### Equações paramétricas, simétricas e geral da reta no plano e no espaço

**Equação vetorial da reta**:(pág.103)

$(x, y, z)= (x_{1}, y_{1}, z_{1}) + t(a, b, c)$


**Exemplo:**A reta $r$ passa pelo ponto A$(1, -1, 4)$ e tem direção do vetor $\vec{v}$=$(2,3,2)$, determine sua equação vetorial.

**Resolução:**A equação vetorial é determinada por $(x, y, z)= (x_{1}, y_{1}, z_{1}) + t(a, b, c)$, onde $(x_{1},y_{1},z_{1})$ é o vetor por onde ela passa e $(a,b,c)$ o ponto que pertence a reta, portando a equação vetorial será:

 $(x, y, z)= (2, 3, 2) + t(1, -1, 4)$



**Equação paramétrica da reta**:(pág.105)

partindo da equação vetorial da reta, $(x, y, z)= (x_{1}, y_{1}, z_{1}) + t(a, b, c)$ conseguimos demonstrar a equação paramétrica da reta.

$(x, y, z)= (x_{1}, y_{1}, z_{1}) + t(a, b, c)$, realizando as operações temos, $(x, y, z)= (x_{1}+at, y_{1}+bt, z_{1}+ct)$, reorganizando


$\left\{ \begin{array}{cl}
x = & \ x_{1} \ +at\\
y= &  \ y_{1} \ +bt \\
z= & \ z_{1} \ +ct
\end{array} \right.$




Encontrando assim as equações paramétricas da reta.


**Exemplo**: A reta $r$ que passa pelo ponto A$(3, -4, 2)$ e é paralela ao vetor $\vec{v}$=$(2,1.-3)$, determine as equações paramétricas.

**Resolução:**As equações paramétricas de uma reta são,$\left\{ \begin{array}{cl}
x = & \ x_{1} \ +at\\
y= &  \ y_{1} \ +bt \\
z= & \ z_{1} \ +ct
\end{array} \right.$, para determinar essa equação precisamos de um ponto e um vetor diretor que indique sua direção, como no problema foi dado que a reta $r$ é paralela ao vetor $\vec{v}$=$(2,1.-3)$ por ser paralelo ele já indica a direção, então usaremos ele e o ponto, na equação que fica:



$\left\{ \begin{array}{cl}
x = & \ 3 \ +2t\\
y= &  \ -4 \ +t \\
z= & \ 2 \ -3t
\end{array} \right.$, essas são as equações paramétricas da reta $r$


**Exemplo**:Dado o ponto A$(2, 3, -4)$ e $\vec{v}$=$(1, -2, 3)$, determine.

**a)**As equações paramétricas da reta $r$ que passa por A$(2, 3, -4)$ e tem direção de $\vec{v}$=$(1, -2, 3)$.

**Resolução:**$\left\{ \begin{array}{cl}
x = & \ 2 \ +t\\
y= &  \ 3 \ -2t \\
z= & \ -4 \ +3t
\end{array} \right.$

**b)**Encontre os dois pontos **B** e **C** de $r$ de parâmetros **t=1** e **t=4**.

**Resolução:**

Para **t=1**, $\left\{ \begin{array}{cl}
x = & \ 2 \ +(1)\\
y= &  \ 3 \ -2(1) \\
z= & \ -4 \ +3(1)
\end{array} \right.$=$B(3, 1, -1)$

Para **t=4**, $\left\{ \begin{array}{cl}
x = & \ 2 \ +(4)\\
y= &  \ 3 \ -2(4) \\
z= & \ -4 \ +3(4)
\end{array} \right.$=$C(6, -5, 8)$


**Equação simétrica da reta**:(pág.108)

partindo das equações paramétricas, $x=x_{1}+at$, $y=y_{1}+bt$ e $z=z_{1}+ct$, isolamos o t em cada uma das equações e chegamos as equações simétricas,

$t=\frac{(x-x_{1})}{a}$



$t=\frac{(y-y_{1})}{b}$



$t=\frac{(z-z_{1})}{c}$

igualando as expreções obtemos:

$\frac{(x-x_{1})}{a} =\frac{(y-y_{1})}{b}$ = $\frac{(z-z_{1})}{c}$


**Exemplo**:A reta que passa pelo ponto A$(3, 0, -5)$ e tem direção do vetor $\vec{v}$=$(2, 2, -1)$, determine suas equações simétricas.

**Resolução:**Sabemos que as equações simétricas são dadas por $\frac{(x-x_{1})}{a} =\frac{(y-y_{1})}{b}$ = $\frac{(z-z_{1})}{c}$, então as equações simétricas:

$\frac{x-3}{2} =\frac{y-0}{2}$ = $\frac{z-5}{-1}$

###Parte computacional equações da reta

```python
#Determinação da equação vetorial da reta:

def equacao_vetorial(ponto, vetor_direcao):

# ponto é uma tupla representando as coordenadas (x1, y1, z1)
# vetor_direcao é uma tupla representando as coordenadas (a, b, c)

# Função para formatar a equação vetorial
    def formatar_equacao(x, y, z):
        return f"(x, y, z) = ({ponto[0]}, {ponto[1]}, {ponto[2]}) + t({vetor_direcao[0]}, {vetor_direcao[1]}, {vetor_direcao[2]})"

# Calculando as coordenadas x, y, z da equação vetorial
    x = f"{ponto[0]} + t * {vetor_direcao[0]}"
    y = f"{ponto[1]} + t * {vetor_direcao[1]}"
    z = f"{ponto[2]} + t * {vetor_direcao[2]}"

    return formatar_equacao(x, y, z)

# Pontos fornecidos
ponto_passagem = (2, 3, 2)
vetor_direcao = (1, -1, 4)

# Calculando e exibindo a equação vetorial da reta:

equacao = equacao_vetorial(ponto_passagem, vetor_direcao)
print("Equação vetorial da reta:")
print(equacao)
```

```python
#Determinação da paramétrica da reta e determinação do coeficiente d:

def equacoes_parametricas(ponto, vetor_direcao):
    # ponto é uma tupla representando as coordenadas (x1, y1, z1)
    # vetor_direcao é uma tupla representando as coordenadas (a, b, c)

    # Equações paramétricas
    x = f"{ponto[0]} + t * {vetor_direcao[0]}"
    y = f"{ponto[1]} + t * {vetor_direcao[1]}"
    z = f"{ponto[2]} + t * {vetor_direcao[2]}"

    return x, y, z

# Pontos fornecidos
ponto_passagem = (2, 3, -4)
vetor_direcao = (1, -2, 3)

# Equações paramétricas da reta
x_param, y_param, z_param = equacoes_parametricas(ponto_passagem, vetor_direcao)
print("a) Equações paramétricas da reta:")
print(f"x = {x_param}, y = {y_param}, z = {z_param}")

def ponto_na_reta(ponto, vetor_direcao, t):
    # Encontrar o ponto na reta para o parâmetro t dado
    x = ponto[0] + t * vetor_direcao[0]
    y = ponto[1] + t * vetor_direcao[1]
    z = ponto[2] + t * vetor_direcao[2]

    return x, y, z

# Pontos fornecidos
ponto_passagem = (2, 3, -4)
vetor_direcao = (1, -2, 3)

# Parâmetro t
t = 1

# Encontrar o ponto na reta para o parâmetro t dado
ponto_t = ponto_na_reta(ponto_passagem, vetor_direcao, t)
print(f"b) Ponto na reta para t = {t}: {ponto_t}")
```

```python
#Determinação das equações simétricas da reta:

def equacoes_simetricas_da_reta(ponto, vetor_diretor):
    # Extrai as coordenadas do ponto e do vetor diretor
    x0, y0, z0 = ponto
    a, b, c = vetor_diretor

# Equações simétricas da reta (fórmula):

# (x - x0) / a = (y - y0) / b = (z - z0) / c

# Realiza o jogo de sinais caso uma das coordenadas seja negativa:

    if a < 0:
        x0 *= -1
        a *= -1
    if b < 0:
        y0 *= -1
        b *= -1
    if c < 0:
        z0 *= -1
        c *= -1

    # Simplifica as equações simétricas
    equacoes_simetricas = []
    if a != 0:
        equacoes_simetricas.append(f"(x - {x0}) / {a}")
    if b != 0:
        equacoes_simetricas.append(f"(y - {y0}) / {b}")
    if c != 0:
        equacoes_simetricas.append(f"(z - {z0}) / {c}")

    return equacoes_simetricas

# Ponto A e vetor diretor v
ponto_A = (3, 0, -5)
vetor_v = (2, 2, -1)

# Calcula as equações simétricas da reta
equacoes_simetricas = equacoes_simetricas_da_reta(ponto_A, vetor_v)

# Imprime as equações simétricas
print("Equações simétricas da reta:")
for i, eq in enumerate(equacoes_simetricas):
    if i != len(equacoes_simetricas) - 1:
        print(eq, end=" = ")
    else:
        print(eq, end="")
```
